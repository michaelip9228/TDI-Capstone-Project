```python
business_df = business_df.dropna(subset=['Latitude', 'Longitude'])
```

```python
def colorfunc_business_type(row): 
    if row["Which type of business is this?"] == 'Food and Beverage': 
        val = "blue"
    elif row["Which type of business is this?"] == 'Retail':
        val = "red"
    else: 
        val = "green"
    return val 

business_df["business type color"] = business_df.apply(colorfunc_business_type, axis = 1) 
```

```python
business_list = business_df[['Name of Business', 
                            'Which type of business is this?', 
                            'Borough',
                            'Latitude', 
                            'Longitude', 
                             'business type color'
                           ]]
```

```python
business_list = business_list.values.tolist()
```
